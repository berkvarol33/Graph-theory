NetworkX tutorial
----------------

----
creating a graph


```python
import networkx as nx
import matplotlib.pyplot as plt
import random
import numpy as np
```

```python
#Create a Graph
G=nx.Graph()
```

add edges

```python
G.add_edge('A','B',weight=1)
G.add_edge('A', 'C', weight=3)
G.add_edge('B', 'C', weight=1)
G.add_edge('B', 'D', weight=2)
G.add_edge('C', 'D', weight=2)
```

draw graph

```python

# Draw the graph
nx.draw(G, with_labels=True)

plt.show()
```

-------------------------
-------------------
Shortest Path Problem
---

find the shortest path between nodes A an B

```python
path = nx.shortest_path(G, 'A', 'D')
print(path)
```

--------------
--------------
now lets plot the shortest path

```python

path_edges = list(zip(path,path[1:]))
print (path_edges)
```

```python
# Draw the graph
pos = nx.spring_layout(G) # to position our graphs
nx.draw(G,pos, with_labels=True) #drawing the graph


# Draw the shortest path on top of the main graph
nx.draw_networkx_nodes(G,pos,nodelist=path,node_color='r')
nx.draw_networkx_edges(G,pos,edgelist=path_edges,edge_color='r',width=10)
```
